### 1.6 기하객체 - 중요

```r
mpg  %>%  ggplot() + geom_point(aes(x= displ, y = hwy))
```

```r
mpg  %>%  ggplot() + geom_smooth(aes(x= displ, y = hwy))
```

```r
mpg  %>%  ggplot() + geom_smooth(aes(x= displ, y = hwy, linetype = drv))
```

```r
mpg  %>%  ggplot(aes(x = displ, y = hwy, color = drv)) + 
    geom_point() + geom_smooth(aes(linetype = drv))
```

```r
mpg  %>%  ggplot(aes(x = displ, y = hwy, group = drv)) + 
    geom_smooth()
```

```r
mpg  %>%  ggplot(aes(x = displ, y = hwy, group = drv)) + 
    geom_smooth(aes(color = drv),
               show.legend = F)
```

```r
mpg  %>% ggplot(aes(x = displ, y = hwy)) + 
    geom_point(aes(color = class)) + 
    geom_smooth( data = mpg %>% filter(class == "subcompact"),
               se = F)
```

```r
mpg  %>% ggplot(aes(x= displ, y = hwy, color = drv)) + 
    geom_point() + geom_smooth(se = F)
```

```r
### 1.6.1 연습문제 6
- 키포인트 : 6번째 플롯을 그리기 위해서는 점의 shape를 바꿔야함
library(grid)
library(gridExtra)
```

```r
p1 = mpg  %>% ggplot(aes(x = displ, y = hwy)) +
  geom_point() + geom_smooth( se = F)

p2 = mpg  %>% ggplot(aes(x = displ, y = hwy)) +
  geom_point() + geom_smooth(aes(group = drv),se = F)

p3 = mpg  %>% ggplot(aes(x = displ, y = hwy, color = drv)) +
  geom_point() + geom_smooth(se = F)

p4 = mpg  %>% ggplot(aes(x = displ, y = hwy)) +
  geom_point(aes(color = drv)) + geom_smooth(se = F)
p5 = mpg  %>% ggplot(aes(x= displ, y = hwy)) + 
  geom_point(aes(color = drv)) + geom_smooth(aes(linetype = drv), se = F)

p6 = mpg  %>% ggplot(aes(x = displ, y = hwy)) + geom_point(aes(fill= drv), 
                                                           stroke = 2, color = "white", shape = 21)

grid.arrange(p1,p2,p3,p4,p5,p6, ncol = 2)
```
