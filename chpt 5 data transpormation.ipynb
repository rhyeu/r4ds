### 3.6 summarize

```r
flights  %>% summarize(delay = mean(dep_delay, na.rm = T))
```

```r
flights   %>% group_by(year, month, day) %>% 
    summarize(delay = mean(dep_delay, na.rm = T))  %>% head()
```

```r
###   
delay <- flights %>% group_by(dest)  %>% 
summarize(count = n(),
         dist = mean(distance, na.rm = T),
         delay = mean(arr_delay, na.rm = T))  %>% 
    filter(count > 20, dest != "HNL")

delay  %>% ggplot(aes(x = dist, y = delay)) + geom_point(aes(size = count), alpha = 0.33) +
    geom_smooth(se = F)
```

#### 3.6.2 

```r
not_cancelled  <- flights  %>% filter(!is.na(dep_delay), !is.na(arr_delay))

not_cancelled  %>% group_by(year, month, day)  %>% 
    summarize(mean = mean(dep_delay))  %>% head()
```

```r
not_cancelled  %>% group_by(tailnum)  %>% 
    summarize(delay = mean(arr_delay))  %>% ggplot(aes(delay)) + geom_freqpoly()
```

```r
not_cancelled  %>% group_by(tailnum)  %>% 
    summarize(delay = mean(arr_delay),
             n = n())  %>% ggplot(aes(x= n, y = delay)) + geom_point(alpha =0.1)
```

```r
not_cancelled  %>% group_by(tailnum)  %>% 
    summarize(delay = mean(arr_delay),
             n = n())  %>% filter(n > 25) %>% ggplot(aes(x= n, y = delay)) + geom_point(alpha =0.1)
```

```r
battings = as_tibble(Lahman::Batting)
batters <- battings  %>% group_by(playerID)  %>% 
    summarize(
    ba = sum(H, na.rm = T) / sum(AB, na.rm = T),
    ab = sum(AB, na.rm = T))

batters  %>% filter(ab>100)  %>% 
ggplot(aes(x = ab, y =ba)) + geom_point() + geom_smooth(se = F)
```
